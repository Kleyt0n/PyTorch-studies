# Dataset Transforms

```python
import torch
import torchvision
from torch.utils.data import Dataset
import numpy as np
```

```python
dataset = torchvision.datasets.MNIST(download=True,
    root='./data', transform = torchvision.transforms.ToTensor())
```

```python
class BostonDataset(Dataset):

  def __init__(self, transform=None):
    #data loading
    xy = np.loadtxt('/content/drive/MyDrive/SummerJobLondon/summer2021/code/PyTorchStudies/tutorial/boston.csv', delimiter=',', dtype=np.float32, skiprows=1)
    self.x = torch.from_numpy(xy[:,1:])
    self.y = torch.from_numpy(xy[:,[0]]) #n_samples, 1
    self.n_samples = xy.shape[0]
  
    self.transform = transform

  def __getitem__(self, index):
    sample = self.x[index], self.y[index]
  
    if self.transform:
      sample = self.transform(sample)

    return sample

  def __len__(self):
    return self.n_samples

class ToTensor:
  def __call__(self, sample):
    inputs, targets = sample 
    return torch.from_numpy(inputs), torch.from_numpy(targets)

class MulTransform:
  def __init__(self, factor):
    self.factor = factor

  def __call__(self, sample):
    inputs, target = sample
    inputs *= self.factor
    return inputs, target

dataset = BostonDataset(transform=None)
first_data = dataset[0]
features, labels = first_data
print(features)
print(type(features), type(labels))

composed = torchvision.transforms.Compose([ToTensor(), MulTransform(2)])
dataset = BostonDataset(transform=None)
first_data = dataset[0]
features, labels = first_data

print(features)
print(type(features), type(labels))
```
